```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
import numpy as np
# Check for NaN
print(df.isna().sum())


# Check for zeros 
print((df == 0).sum())
```

```python
columns_with_nan = [
    'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered',
    'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered',
    '1 week change', '1 week % increase'
]

for col in columns_with_nan:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].median())
```

```python
def remove_outliers(df,column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df = remove_outliers(df, 'Confirmed')
df = remove_outliers(df, 'Deaths')
df = remove_outliers(df, 'Recovered')
df = remove_outliers(df, 'Active')
df = remove_outliers(df, 'New cases')
df = remove_outliers(df, 'New deaths')
df = remove_outliers(df, 'New recovered')
df = remove_outliers(df, 'Deaths / 100 Cases')
df = remove_outliers(df, 'Recovered / 100 Cases')
df = remove_outliers(df, 'Confirmed last week')
df = remove_outliers(df, '1 week change')
df = remove_outliers(df, '1 week % increase')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df['Deaths'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['Deaths'], bins=30, kde=True, color='red', alpha=0.8)
plt.xlabel("Number of Deaths")
plt.ylabel("Number of Countries")
plt.title("Distribution of Deaths Across Countries")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[["Confirmed","Deaths","Recovered","Active"]].corr(), annot=True)
plt.title("HEATMAP")
plt.show()
```

```python

top_confirmed = df.groupby("Country/Region")["Confirmed"].sum().sort_values(ascending=False).head(15)
top_confirmed.plot(kind="bar", figsize=(12, 6), color="skyblue", edgecolor="black")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")
plt.title("Top 15 Countries by Confirmed COVID-19 Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
